```python
Principal = float(input("Enter your value for the principal: "))
Rate = float(input("Enter your Rate: "))
No_of_Compounded_Interest=float(input("Enter no of years the interest compounded: "))
Time=float(input("Enter your number of years: "))
Amount=(Principal*(1+(Rate/No_of_Compounded_Interest)))**(No_of_Compounded_Interest*Time)
print(Amount)
```

```python
P_M_T = float(input("Enter your value for the amount of money deposited each period: "))
Rate = float(input("Enter your Rate: "))
No_of_Compounded_Interest=float(input("Enter no of years the interest compounded: "))
Time=float(input("Enter your number of years: "))

periodic_rate=(Rate/No_of_Compounded_Interest)
total_periods=No_of_Compounded_Interest*Time

Amount=P_M_T*(((1+periodic_rate)**total_periods)-1)/periodic_rate
print(Amount)
```
