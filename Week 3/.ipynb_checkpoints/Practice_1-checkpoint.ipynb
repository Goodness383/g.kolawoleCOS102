```python
import tkinter as tk 

window= tk.Tk()
window.title("Employees Register")
window.geometry("400x400")
window.config(background="#345beb")

#icon=PhotoImage(file="logo.jpg")
#icon.iconphoto(True,icon)


username_label = tk.Label(window, text="Username", font=("Times New Roman",20,"bold"))
username_label.pack(pady=5)

username_entry = tk.Entry(window)
username_entry.pack(pady=5)

password_label = tk.Label(window, text="Password"  font=("Times New Roman",20,"bold"))
password_label.pack(pady=5)

password_entry = tk.Entry(window, show="*")
password_entry.pack(pady=5)

button = tk.Button(
    window,
    text="Click Me",
    bg="White",
    fg="Black",
    font=("Arial", 14),
    width=15,
    height=2)
button.pack(pady=5)

   




window.mainloop()
```
