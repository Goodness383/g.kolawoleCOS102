```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# Load the dataset
try:
    df = pd.read_csv('glogistics.csv')
    # Cleaning column names just in case there are hidden spaces
    df.columns = df.columns.str.strip()
except FileNotFoundError:
    messagebox.showerror("Error", "CSV file not found!")
    df = pd.DataFrame() 

def verify_employee():
    name_input = entry_name.get().strip()
    dept_input = entry_dept.get().strip()

    # Search for the employee (case-insensitive)
    match = df[(df['Name'].str.lower() == name_input.lower()) & 
               (df['Department'].str.lower() == dept_input.lower())]

    if not match.empty:
        # Get list of colleagues in the same department
        colleagues = df[df['Department'].str.lower() == dept_input.lower()]['Name'].tolist()
        colleagues.remove(match.iloc[0]['Name']) # Remove the user from their own colleague list
        
        colleague_str = ", ".join(colleagues) if colleagues else "No other members."
        
        messagebox.showinfo("Success", 
                            f"Welcome, {match.iloc[0]['Name']}!\n\n"
                            f"Your Department Members: \n{colleague_str}")
    else:
        messagebox.showerror("Access Denied", "Employee does not exist in this department.")

# --- GUI Setup ---
root = tk.Tk()
root.title("GIG Logistics - Name Verification")
root.geometry("400x250")

tk.Label(root, text="Employee Name:", font=("Arial", 10)).pack(pady=5)
entry_name = tk.Entry(root, width=30)
entry_name.pack()

tk.Label(root, text="Department:", font=("Arial", 10)).pack(pady=5)
entry_dept = tk.Entry(root, width=30)
entry_dept.pack()

btn_verify = tk.Button(root, text="Verify Identity", command=verify_employee, bg="navy", fg="white")
btn_verify.pack(pady=20)

root.mainloop()
```
